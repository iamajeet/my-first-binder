```python
import pandas as pd
weather_data = {
    'day': ['1/1/2017','1/2/2017','1/3/2017','1/4/2017','1/5/2017','1/6/2017'],
    'temperature': [32,35,28,24,32,31],
    'windspeed': [6,7,2,7,4,2],
    'event': ['Rain', 'Sunny', 'Snow','Snow','Rain', 'Sunny']
}
```

```python
df = pd.DataFrame(weather_data)
df = pd.read_csv("weather.csv")
df
```

### fillna

```python
new_df= df.fillna(0)
new_df
```

```python
#what does the 0 mean in event? so replacing all na with single value is not good geuss
new_df= df.fillna({"temperature":0,
                   "windspeed":0,
                   "event":"no event"
})
new_df
```

```python
#temparature 0 is not good geuss , it will hit bad while taking median
new_df= df.fillna(method="ffill")
new_df
```

```python
new_df= df.fillna(method="bfill")
new_df
```

```python
#along column bfill
new_df= df.fillna(method="bfill", axis="columns") 
#or
#new_df= df.fillna(method="bfill", axis=1)
new_df
```

```python
new_df= df.fillna(method="ffill",limit=1)
new_df
```

```python
df
```

```python
#by default it is linear interpolate
new_df= df.interpolate()
new_df
```

```python
#we are doing interpolate by time
new_df= df.interpolate(method="time")
new_df
```

```python
import numpy as np
df = pd.read_csv("weather2.csv")
df
```

```python
new_df= df.replace(-99999,np.NaN)
new_df
```

```python
#replace multiple values
new_df= df.replace([-99999, 5.0],np.NaN)
new_df
```

```python
#better replace multiple values
new_df= df.replace({"temperature":-99999,
                    "event":"0"
},np.NaN)
new_df
```

```python
#Replacing by using mapping
new_df = df.replace({
        -99999: np.nan,
        '0': 'Sunny',
    })
new_df
```

```python
#Regex
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace('[A-Za-z]','', regex=True) 
new_df
```

```python
#Regex better way
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```

```python
#Pandas Concatenate Tutorial
#Basic Concatenation
import pandas as pd

india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
df
```

```python
#Ignore Index
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
#Concatenation And Keys
df = pd.concat([india_weather, us_weather], keys=["india", "us"])
df
```

```python
df.loc["us"]
```

```python
df.loc["india"]
```

```python
#Concatenation Using Index
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
df = pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#Concatenate dataframe with series
s = pd.Series(["Humid","Dry","Rain"], name="event")
s
```

```python
df = pd.concat([temperature_df,s],axis=1)
df
```
